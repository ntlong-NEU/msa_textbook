## 8.1. Biến ẩn (Latent Variables): Linh hồn của cấu trúc đa biến

Trong các chương trước, chúng ta chủ yếu làm việc với các biến định lượng có thể quan sát và đo lường trực tiếp (Observed/Manifest Variables) như: Doanh thu, Lợi nhuận, Giá vàng, hay Dư nợ. Tuy nhiên, trong thế giới Kinh doanh và Marketing thực tế, những khái niệm quan trọng nhất lại thường là những thực thể **ẩn** (Latent Constructs).

### 8.1.1. Khái niệm Biến ẩn (Latent Variables)

**Biến ẩn (Latent Variable)** là một đặc tính hoặc trạng thái tâm lý không thể quan sát hoặc đo lường trực tiếp bằng một con số hay một dụng cụ đo đơn nhất. Chúng ta chỉ có thể suy diễn sự tồn tại và mức độ của nó thông qua một tập hợp các biểu hiện bên ngoài (Indicators) hoặc các câu hỏi khảo sát.

**Ví dụ tiêu biểu trong Kinh tế và Quản trị:**
1. **Sự hài lòng của khách hàng (Customer Satisfaction):** Bạn không thể dùng thước đo để biết khách hàng hài lòng bao nhiêu "cm". Thay vào đó, bạn đặt ra 5 câu hỏi (Likert 1-5): "Bạn có quay lại không?", "Bạn có giới thiệu cho bạn bè không?", "Dịch vụ có nhanh không?",... Tập hợp các câu trả lời này chính là những mảnh ghép phản chiếu biến ẩn "Sự hài lòng".
2. **Lòng trung thành thương hiệu (Brand Loyalty):** Đây là một cấu trúc tâm lý phức tạp bao gồm cả hành vi (mua lặp lại) và thái độ (sự gắn kết cảm xúc).
3. **Năng lực lãnh đạo (Leadership Capability):** Một biến ẩn quan trọng trong quản trị nhân sự, được đo qua khả năng truyền cảm hứng, tầm nhìn chiến lược, và kỹ năng ra quyết định.
4. **Chỉ số IQ (Intelligence Quotient):** Trí thông minh là biến ẩn; điểm số các bài thi toán, ngôn ngữ, logic là các biến quan sát.

### 8.1.2. Triết lý của phép đo (Measurement Philosophy)

Trong phân tích đa biến, chúng ta chấp nhận một giả định triết học quan trọng: **Biến ẩn là nguyên nhân gây ra sự biến thiên của các biến quan sát**. 

Hãy tưởng tượng biến ẩn là một thực thể nằm trong bóng tối và chiếu ánh sáng ra ngoài. Các biến quan sát là những hình bóng (shadows) in lên tường. Bằng cách phân tích hình thù và sự chuyển động của những chiếc bóng đó, chúng ta có thể dựng lại hình ảnh của thực thể gốc. Nếu 5 câu hỏi khảo sát về dịch vụ ngân hàng đều có điểm số tăng lên, ta suy luận rằng "Chất lượng dịch vụ" (biến ẩn) đang thực sự cải thiện.

## 8.2. Mô hình Toán học EFA và Phân biệt EFA với PCA

Phân tích Nhân tố Khám phá (EFA) là một kỹ thuật thống kê dùng để xác định cấu trúc ẩn giấu của một tập hợp các biến quan sát, bằng cách tìm ra một số ít các nhân tố chung giải thích cho mối tương quan giữa chúng.

### 8.2.1. Phương trình cơ bản của EFA

Giả sử ta có $p$ biến quan sát $X_1, X_2, \dots, X_p$. Mô hình EFA giả định rằng mỗi biến $X_j$ là một tổ hợp tuyến tính của $m$ nhân tố chung ($F$) và một thành phần sai số đặc thù ($\epsilon$). Phương trình cho biến $j$ trên đối tượng $i$ là:

$$x_{ij} = \mu_j + \lambda_{j1}f_{i1} + \lambda_{j2}f_{i2} + \dots + \lambda_{jm}f_{im} + \epsilon_{ij}$$

Viết dưới dạng ma trận cho toàn bộ hệ thống:

$$\mathbf{X} = \boldsymbol{\mu} + \boldsymbol{\Lambda} \boldsymbol{f} + \boldsymbol{\epsilon}$$

**Giải thích chi tiết các ký hiệu toán học:**
- $\mathbf{X}$: Ma trận dữ liệu quan sát kích thước $N \times p$.
- $\boldsymbol{\mu}$: Vector kỳ vọng (trung bình) của các biến.
- $\boldsymbol{f} = [f_1, f_2, \dots, f_m]^T$: Vector chứa $m$ **Nhân tố chung (Common Factors)**.
- $\boldsymbol{\Lambda} = [\lambda_{jk}]$: Ma trận **Factor Loadings** kích thước $p \times m$.
- $\boldsymbol{\epsilon}$: Vector chứa các **Unique Factors/Error**.
  1. **Specific variance (Phương sai riêng):** Những thứ chỉ thuộc về biến đó.
  2. **Measurement error (Sai số đo lường):** Nhiễu trong quá trình thu thập dữ liệu.

---


### 8.2.2. Phân tách Phương sai (Variance Partitioning)

Trong EFA, tổng phương sai của một biến chuẩn hóa ($Var(X_j)=1$) được chia làm 2 phần:

$$1 = h_j^2 + u_j^2$$

1. **$h_j^2$ (Communality - Độ chung):** Tổng lượng phương sai của biến $X_j$ được giải thích bởi tất cả các nhân tố chung. 
   $$h_j^2 = \sum_{k=1}^{m} \lambda_{jk}^2$$
2. **$u_j^2$ (Uniqueness - Độ riêng):** Lượng phương sai riêng biệt và sai số của biến đó ($u_j^2 = 1 - h_j^2$).

---

### 8.2.3. Phân biệt triết lý: EFA vs. PCA (Rất quan trọng)

Mặc dù cả hai đều dùng để giảm chiều dữ liệu, EFA và PCA đại diện cho hai trường phái tư duy hoàn toàn khác biệt.

| Tiêu chí | PCA (Principal Component Analysis) | EFA (Exploratory Factor Analysis) |
| :--- | :--- | :--- |
| **Mục đích** | **Tổng hợp thông tin (Data Reduction).** Gom nhiều biến thành ít thành phần để dễ xử lý. | **Tìm cấu trúc ẩn (Latent Structure).** Tìm các thực thể ẩn giấu gây ra tương quan giữa các biến. |
| **Hướng tác động** | Biến gốc tạo ra Thành phần chính ($X \to PC$). | Nhân tố chung tạo ra Biến gốc ($F \to X$). |
| **Giả định sai số** | **Không có sai số.** Giả định toàn bộ phương sai của dữ liệu là có ích và được đưa vào tính toán. | **Tách bạch sai số.** Chỉ quan tâm đến phần phương sai chung (Communality). Loại bỏ nhiễu và phương sai đặc thù. |
| **Toán học** | PC là tổ hợp tuyến tính của X ($PC = w^T X$). | X là tổ hợp tuyến tính của F ($X = \Lambda F + e$). |
| **Ứng dụng** | Khi bạn muốn dự báo hoặc làm sạch đa cộng tuyến (PCR). | Khi bạn muốn xây dựng thang đo, kiểm định lý thuyết tâm lý/kinh tế. |

**Tóm lại:** PCA là phép toán tóm tắt (Synthesis), còn EFA là phép toán phân tích căn nguyên (Analysis of causes).

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import FactorAnalysis, PCA
from sklearn.preprocessing import StandardScaler

# ========================================================
# BƯỚC 1: GIẢ LẬP DỮ LIỆU KHẢO SÁT MARKETING (N=500)
# ========================================================
# Chúng ta giả lập 6 câu hỏi đo lường 2 biến ẩn:
# Nhân tố 1: Trải nghiệm Dịch vụ (Service - S1, S2, S3)
# Nhân tố 2: Chất lượng Sản phẩm (Product - P1, P2, P3)

np.random.seed(42)
N = 500

# Sinh 2 nhân tố ẩn độc lập (F1, F2)
F1_Service = np.random.normal(0, 1, N)
F2_Product = np.random.normal(0, 1, N)

# Tạo các biến quan sát (Manifest variables) với hệ số tải và sai số riêng
S1 = 0.8 * F1_Service + np.random.normal(0, 0.4, N)
S2 = 0.7 * F1_Service + np.random.normal(0, 0.5, N)
S3 = 0.9 * F1_Service + np.random.normal(0, 0.3, N)

P1 = 0.85 * F2_Product + np.random.normal(0, 0.4, N)
P2 = 0.75 * F2_Product + np.random.normal(0, 0.5, N)
P3 = 0.65 * F2_Product + np.random.normal(0, 0.6, N)

df = pd.DataFrame({'S1': S1, 'S2': S2, 'S3': S3, 'P1': P1, 'P2': P2, 'P3': P3})

# Chuẩn hóa dữ liệu
X = StandardScaler().fit_transform(df)

# ========================================================
# BƯỚC 2: THỰC THI EFA (FACTOR ANALYSIS)
# ========================================================
fa = FactorAnalysis(n_components=2, random_state=42)
fa.fit(X)

# Trích xuất ma trận loadings
fa_loadings = pd.DataFrame(fa.components_.T, columns=['Factor1', 'Factor2'], index=df.columns)

print("---- MA TRẬN HỆ SỐ TẢI (FACTOR LOADINGS) ----")
display(fa_loadings.round(3))

# Tính Uniqueness (Độ riêng)
uniqueness = fa.noise_variance_
print("\n---- PHƯƠNG SAI RIÊNG (UNIQUENESS) ----")
print(pd.Series(uniqueness, index=df.columns).round(3))
```

### Phân tích diễn giải Output EFA:

1. **Ma trận Loadings:** 
   - Bạn có thể thấy `S1, S2, S3` có hệ số tải rất cao ở **Factor 1** (~0.8-0.9) và gần bằng 0 ở Factor 2. Điều này chứng minh Factor 1 chính là biến ẩn "Trải nghiệm Dịch vụ".
   - Ngược lại, `P1, P2, P3` tải mạnh vào **Factor 2**. Đây là bằng chứng thực nghiệm cho thấy cấu trúc lý thuyết 2 nhân tố ban đầu là chính xác.
2. **Uniqueness (Độ riêng):** 
   - Giá trị `0.16` của `S3` cho thấy chỉ có 16% biến thiên của câu hỏi S3 là nhiễu hoặc đặc điểm riêng. 84% còn lại là do nhân tố "Dịch vụ" chi phối (Độ chung Communality cao). 
   - Biến nào có Uniqueness quá cao (> 0.5) thường bị coi là biến rác trong phân tích thang đo vì nó không phản ánh tốt biến ẩn.

## 8.3. Kiểm định sự phù hợp của dữ liệu trong EFA

Trước khi tiến hành trích xuất nhân tố, nhà nghiên cứu bắt buộc phải chứng minh rằng các biến quan sát trong ma trận dữ liệu $\mathbf{X}$ có mối tương quan đủ mạnh để phản chiếu một cấu trúc ẩn chung. Nếu các biến hoàn toàn độc lập, việc tìm kiếm nhân tố là vô nghĩa. Hai công cụ chẩn đoán kinh điển cho vấn đề này là chỉ số KMO và kiểm định Bartlett.

---

### 8.3.1. Chỉ số KMO (Kaiser-Meyer-Olkin Measure of Sampling Adequacy)

**Bản chất:** Chỉ số KMO đo lường mức độ phù hợp của mẫu bằng cách so sánh cường độ của các hệ số tương quan quan sát được ($r_{ij}$) với các hệ số tương quan riêng ($a_{ij}$). 

**Công thức toán học:**

$$KMO = \frac{\sum_{i \neq j} r_{ij}^2}{\sum_{i \neq j} r_{ij}^2 + \sum_{i \neq j} a_{ij}^2}$$

**Giải thích ký hiệu:**
- $r_{ij}$: Hệ số tương quan Pearson giữa biến $X_i$ và $X_j$. 
- $a_{ij}$: Hệ số tương quan riêng (Partial correlation) giữa biến $X_i$ và $X_j$ sau khi đã loại bỏ ảnh hưởng của tất cả các biến còn lại trong mô hình.

**Triết lý thống kê:** 
- Nếu dữ liệu thực sự có cấu trúc nhân tố ẩn, các biến quan sát sẽ tương quan mạnh với nhau ($r_{ij}^2$ lớn), đồng thời các tương quan riêng $a_{ij}^2$ (phần tương quan không giải thích được bởi nhân tố chung) sẽ rất nhỏ. 
- Khi đó, mẫu số sẽ xấp xỉ tử số, và $KMO$ sẽ tiến gần đến $1.0$.

**Ngưỡng chấp nhận (Thresholds):**
- $KMO \ge 0.9$: Rất tốt (Marvelous).
- $0.8 \le KMO < 0.9$: Tốt (Meritorious).
- $0.7 \le KMO < 0.8$: Đạt yêu cầu (Middling).
- $0.6 \le KMO < 0.7$: Tạm chấp nhận được (Mediocre).
- $0.5 \le KMO < 0.6$: Rất kém (Miserable).
- $KMO < 0.5$: Không thể thực hiện EFA (Unacceptable).

---

### 8.3.2. Kiểm định Bartlett về tính cầu (Bartlett's Test of Sphericity)

**Mục tiêu:** Kiểm tra giả thuyết cho rằng ma trận tương quan $\mathbf{R}$ của dữ liệu mẫu là một ma trận đơn vị $\mathbf{I}$ (Identity matrix).

**Giả thuyết:**
- $H_0: \mathbf{R} = \mathbf{I}$. (Các biến hoàn toàn không tương quan tuyến tính với nhau).
- $H_A: \mathbf{R} \neq \mathbf{I}$. (Tồn tại ít nhất một cặp biến có tương quan đáng kể).

**Toán học:** Thống kê kiểm định Chi-bình phương ($\chi^2$) được tính dựa trên định thức của ma trận tương quan $|\mathbf{R}|$: 

$$\chi^2 = -\left[ n - 1 - \frac{2p + 5}{6} \right] \ln |\mathbf{R}|$$

**Giải thích ký hiệu:**
- $n$: Số lượng quan sát (mẫu).
- $p$: Số lượng biến quan sát.
- $|\mathbf{R}|$: Định thức của ma trận tương quan. Nếu $|\mathbf{R}| = 1$ (ma trận đơn vị), thì $\ln(1) = 0$, dẫn đến $\chi^2 = 0$.

**Kết luận:** Nếu giá trị $p$-value $< 0.05$, chúng ta bác bỏ $H_0$. Điều này chứng minh ma trận dữ liệu không phải là "hình cầu" (tức các biến không độc lập), và việc thực hiện phân tích đa biến là có cơ sở khoa học.

## 8.4. Xoay nhân tố (Factor Rotation): Nghệ thuật Giải mã

Sau khi trích xuất nhân tố (thường bằng phương pháp Trục chính - Principal Axis Factoring), các hệ số tải (Loadings) ban đầu thường rất khó giải thích. Một biến có thể tải mạnh vào nhiều nhân tố đồng thời (Cross-loadings), gây ra sự mơ hồ trong việc định danh bản chất nhân tố.

### 8.4.1. Tại sao phải xoay?

Mục tiêu của xoay nhân tố là đạt được **Cấu trúc đơn giản (Simple Structure)** theo tiêu chuẩn của Thurstone (1947):
1. Mỗi biến quan sát chỉ nên có hệ số tải cao trên **duy nhất một** nhân tố.
2. Mỗi nhân tố nên có hệ số tải cao cho một nhóm biến quan sát nhất định và hệ số tải gần bằng 0 cho các biến còn lại.

Về mặt hình học, xoay nhân tố là việc giữ nguyên vị trí của các điểm dữ liệu nhưng **xoay các trục tọa độ (nhân tố)** sao cho chúng đi sát qua các cụm điểm dữ liệu nhất. 

---

### 8.4.2. Xoay vuông góc (Orthogonal Rotation - Varimax)

**Triết lý:** Các nhân tố sau khi xoay vẫn duy trì góc $90^\circ$ với nhau. Điều này ngụ ý rằng các nhân tố ẩn là **hoàn toàn độc lập** và không tương quan.

**Thuật toán Varimax (Variance Maximization):** Đây là phương pháp phổ biến nhất. Nó tìm cách tối đa hóa phương sai của các bình phương hệ số tải trong mỗi cột của ma trận nhân tố. 
- Nó ép các hệ số tải lớn trở nên lớn hơn và các hệ số tải nhỏ tiến về 0.
- **Ưu điểm:** Kết quả cực kỳ rõ ràng, dễ tách biệt các biến vào các nhân tố khác nhau.

### 8.4.3. Xoay xiên (Oblique Rotation - Promax/Oblimin)

**Triết lý:** Các trục nhân tố không nhất thiết phải vuông góc. Nó cho phép các nhân tố ẩn **tương quan** với nhau.

**Tại sao dùng trong Kinh tế/Hành vi?** 
Trong thực tế, các khái niệm ẩn hiếm khi độc lập. Ví dụ: "Sự hài lòng" và "Lòng trung thành" chắc chắn có mối liên hệ hữu cơ. Việc ép chúng phải vuông góc (Varimax) là một sự áp đặt toán học khiên cưỡng và có thể làm sai lệch bản chất dữ liệu.

**Phân tích ma trận đầu ra:** Trong xoay xiên, chúng ta có 2 ma trận quan trọng:
1. **Ma trận Cấu trúc (Structure Matrix):** Chứa tương quan giữa biến và nhân tố (bao gồm cả ảnh hưởng gián tiếp từ các nhân tố khác).
2. **Ma trận Mẫu (Pattern Matrix):** Chỉ chứa các hệ số tác động trực tiếp của nhân tố lên biến. Đây là ma trận quan trọng nhất để diễn giải ý nghĩa nhân tố trong xoay xiên.

---

```python
import numpy as np
import pandas as pd
from factor_analyzer import FactorAnalyzer
from factor_analyzer.factor_analyzer import calculate_kmo, calculate_bartlett_sphericity
from sklearn.preprocessing import StandardScaler

# ========================================================
# BƯỚC 1: TẠO DỮ LIỆU KHẢO SÁT MARKETING (N=400)
# 9 biến đo lường 3 nhân tố: Giá cả (P), Chất lượng (Q), Dịch vụ (S)
# ========================================================
np.random.seed(42)
N = 400

# Tạo 3 nhân tố ẩn có tương quan với nhau (Oblique structure)
# Ma trận tương quan giữa các nhân tố ẩn
factor_corr = np.array([[1.0, 0.4, 0.3], 
                        [0.4, 1.0, 0.5], 
                        [0.3, 0.5, 1.0]])
factors = np.random.multivariate_normal([0,0,0], factor_corr, N)

F_Price, F_Quality, F_Service = factors[:, 0], factors[:, 1], factors[:, 2]

# Tạo 9 biến quan sát
data = {
    'P1': 0.8*F_Price + np.random.normal(0, 0.4, N), 'P2': 0.7*F_Price + np.random.normal(0, 0.4, N), 'P3': 0.85*F_Price + np.random.normal(0, 0.4, N),
    'Q1': 0.75*F_Quality + np.random.normal(0, 0.4, N), 'Q2': 0.8*F_Quality + np.random.normal(0, 0.4, N), 'Q3': 0.7*F_Quality + np.random.normal(0, 0.4, N),
    'S1': 0.9*F_Service + np.random.normal(0, 0.4, N), 'S2': 0.85*F_Service + np.random.normal(0, 0.4, N), 'S3': 0.8*F_Service + np.random.normal(0, 0.4, N)
}
df = pd.DataFrame(data)

# ========================================================
# BƯỚC 2: KIỂM ĐỊNH SỰ PHÙ HỢP
# ========================================================
chi_square_value, p_value = calculate_bartlett_sphericity(df)
kmo_all, kmo_model = calculate_kmo(df)

print(f"---- KIỂM ĐỊNH SỰ PHÙ HỢP ----")
print(f"Bartlett's Test p-value: {p_value:.4f} ({'Đạt' if p_value < 0.05 else 'Không đạt'})")
print(f"Chỉ số KMO tổng thể   : {kmo_model:.4f} ({'Rất tốt' if kmo_model > 0.8 else 'Tạm được'})")

# ========================================================
# BƯỚC 3: THỰC HIỆN XOAY NHÂN TỐ
# ========================================================
# 1. Xoay vuông góc (Varimax)
fa_varimax = FactorAnalyzer(n_factors=3, rotation="varimax")
fa_varimax.fit(df)

# 2. Xoay xiên (Promax)
fa_promax = FactorAnalyzer(n_factors=3, rotation="promax")
fa_promax.fit(df)

print("\n---- MA TRẬN HỆ SỐ TẢI (PATTERN MATRIX - PROMAX) ----")
loadings_df = pd.DataFrame(fa_promax.loadings_, 
                           columns=['Factor_Price', 'Factor_Quality', 'Factor_Service'], 
                           index=df.columns)
display(loadings_df.round(3))
```

## 8.3. Kiểm định sự phù hợp của dữ liệu trong EFA

Trước khi tiến hành trích xuất nhân tố, nhà nghiên cứu bắt buộc phải chứng minh rằng các biến quan sát trong ma trận dữ liệu $\mathbf{X}$ có mối tương quan đủ mạnh để phản chiếu một cấu trúc ẩn chung. Nếu các biến hoàn toàn độc lập, việc tìm kiếm nhân tố là vô nghĩa. Hai công cụ chẩn đoán kinh điển cho vấn đề này là chỉ số KMO và kiểm định Bartlett.

---

### 8.3.1. Chỉ số KMO (Kaiser-Meyer-Olkin Measure of Sampling Adequacy)

**Bản chất:** Chỉ số KMO đo lường mức độ phù hợp của mẫu bằng cách so sánh cường độ của các hệ số tương quan quan sát được ($r_{ij}$) với các hệ số tương quan riêng ($a_{ij}$). 

**Công thức toán học:**

$$KMO = \frac{\sum_{i \neq j} r_{ij}^2}{\sum_{i \neq j} r_{ij}^2 + \sum_{i \neq j} a_{ij}^2}$$

**Giải thích chi tiết các ký hiệu:**
- $r_{ij}$: Hệ số tương quan Pearson (Simple correlation) giữa biến $X_i$ và $X_j$. Đại diện cho tổng lượng biến thiên chung.
- $a_{ij}$: Hệ số tương quan riêng (Partial correlation) giữa biến $X_i$ và $X_j$ sau khi đã loại bỏ ảnh hưởng của tất cả các biến còn lại. Đây chính là các phần tử ngoài đường chéo chính của Ma trận tương quan phản chiếu (Anti-image correlation matrix).

**Triết lý thống kê:** 
- Nếu dữ liệu thực sự có cấu trúc nhân tố ẩn, các biến quan sát sẽ tương quan mạnh với nhau ($r_{ij}^2$ lớn), đồng thời các tương quan riêng $a_{ij}^2$ (phần tương quan không giải thích được bởi nhân tố chung) sẽ rất nhỏ. 
- Khi đó, mẫu số sẽ xấp xỉ tử số, và $KMO$ sẽ tiến gần đến $1.0$.

**Ngưỡng chấp nhận (Thresholds):**
- $KMO \ge 0.9$: Rất tốt (Marvelous).
- $0.8 \le KMO < 0.9$: Tốt (Meritorious).
- $0.7 \le KMO < 0.8$: Đạt yêu cầu (Middling).
- $0.6 \le KMO < 0.7$: Tạm chấp nhận được (Mediocre).
- $0.5 \le KMO < 0.6$: Rất kém (Miserable).
- $KMO < 0.5$: Không thể thực hiện EFA (Unacceptable).

---

### 8.3.2. Kiểm định Bartlett về tính cầu (Bartlett's Test of Sphericity)

**Mục tiêu:** Kiểm tra giả thuyết cho rằng ma trận tương quan $\mathbf{R}$ của dữ liệu mẫu là một ma trận đơn vị $\mathbf{I}$ (Identity matrix).

**Giả thuyết:**
- $H_0: \mathbf{R} = \mathbf{I}$ (Các biến hoàn toàn không tương quan tuyến tính với nhau, ma trận có hình cầu).
- $H_A: \mathbf{R} \neq \mathbf{I}$ (Tồn tại ít nhất một cặp biến có tương quan đáng kể, ma trận không phải hình cầu).

**Toán học:** Thống kê kiểm định Chi-bình phương ($\chi^2$) được tính dựa trên định thức của ma trận tương quan $|\mathbf{R}|$: 

$$\chi^2 = -\left[ n - 1 - \frac{2p + 5}{6} \right] \ln |\mathbf{R}|$$

**Giải thích ký hiệu:**
- $n$: Số lượng quan sát (Sample size).
- $p$: Số lượng biến quan sát.
- $|\mathbf{R}|$: Định thức (Determinant) của ma trận tương quan. 
  - Nếu $|\mathbf{R}| = 1$ (ma trận đơn vị), thì $\ln(1) = 0$, dẫn đến $\chi^2 = 0$. Ta không thể bác bỏ $H_0$.
  - Nếu $|\mathbf{R}|$ tiến về $0$ (tương quan rất mạnh), $\ln|\mathbf{R}|$ sẽ tiến về âm vô cùng, làm $\chi^2$ rất lớn.

**Kết luận thực tiễn:** Nếu giá trị $p$-value $< 0.05$, chúng ta bác bỏ $H_0$. Điều này chứng minh rằng dữ liệu có cấu trúc tương quan đủ để tiến hành phân tích đa biến.

---

## 8.4. Xoay nhân tố (Factor Rotation): Nghệ thuật Giải mã ý nghĩa

Sau khi trích xuất nhân tố ban đầu (Initial extraction), các hệ số tải (Loadings) thường rất khó giải thích. Hiện tượng phổ biến là một biến tải mạnh vào nhiều nhân tố đồng thời (Cross-loadings), hoặc nhân tố đầu tiên nắm giữ quá nhiều biến thiên làm lu mờ các nhân tố sau.

### 8.4.1. Tại sao phải xoay? Để đạt được "Cấu trúc đơn giản" (Simple Structure)

Mục tiêu của xoay nhân tố là đạt được tiêu chuẩn của Thurstone (1947) về một cấu trúc dễ diễn giải:
1. Mỗi biến quan sát chỉ nên có hệ số tải cao trên **duy nhất một** nhân tố.
2. Mỗi nhân tố nên có hệ số tải cao cho một nhóm biến nhất định và hệ số tải gần bằng 0 cho các biến còn lại.

Về mặt hình học, xoay nhân tố là việc giữ nguyên vị trí của các điểm dữ liệu nhưng **xoay các trục nhân tố** sao cho chúng đi sát qua các cụm điểm dữ liệu nhất.

---

### 8.4.2. Xoay vuông góc (Orthogonal Rotation - Varimax)

**Triết lý:** Các nhân tố sau khi xoay vẫn duy trì góc $90^\circ$ với nhau. Điều này ngụ ý rằng các nhân tố ẩn là **hoàn toàn độc lập** và không tương quan ($Cov(F_1, F_2) = 0$).

**Thuật toán Varimax (Variance Maximization):** Đây là phương pháp phổ biến nhất. Nó tìm cách tối đa hóa phương sai của các bình phương hệ số tải trong mỗi cột của ma trận nhân tố. 
- Nó đẩy các hệ số tải lớn trở nên lớn hơn (tiến về 1) và các hệ số tải nhỏ tiến về 0.
- **Toán học:** Cực đại hóa hàm $V = \sum (\lambda_{jk}^2 - \bar{\lambda}_k^2)^2$.

**Khi nào dùng:** Dùng khi lý thuyết cho rằng các khái niệm ẩn là độc lập (ví dụ: "Chiều cao" và "Trí thông minh").

### 8.4.3. Xoay xiên (Oblique Rotation - Promax/Oblimin)

**Triết lý:** Các trục nhân tố không nhất thiết phải vuông góc. Nó cho phép các nhân tố ẩn **tương quan** với nhau.

**Tại sao dùng trong Kinh tế & Hành vi?** 
Trong thực tế, các khái niệm xã hội hiếm khi độc lập. Ví dụ: "Sự hài lòng" và "Lòng trung thành" chắc chắn có mối liên hệ hữu cơ. Việc ép chúng phải vuông góc (Varimax) là một sự áp đặt toán học khiên cưỡng và có thể làm sai lệch bản chất dữ liệu.

**Phân tích ma trận đầu ra:** Trong xoay xiên, chúng ta phải đọc **Ma trận Mẫu (Pattern Matrix)**. 
- **Pattern Matrix:** Chứa các hệ số tác động trực tiếp của nhân tố lên biến (giống hệ số hồi quy chuẩn hóa). Đây là ma trận dùng để đặt tên nhân tố.
- **Structure Matrix:** Chứa tương quan đơn thuần giữa biến và nhân tố (bao gồm cả ảnh hưởng gián tiếp).

---

```python
import numpy as np
import pandas as pd
from factor_analyzer import FactorAnalyzer
from factor_analyzer.factor_analyzer import calculate_kmo, calculate_bartlett_sphericity
from sklearn.preprocessing import StandardScaler

# ========================================================
# BƯỚC 1: TẠO DỮ LIỆU GIẢ LẬP KHẢO SÁT NGÂN HÀNG (N=500)
# 9 biến đo lường 3 nhân tố ẩn: Sự tin tưởng (T), Tính tiện lợi (C), Phí dịch vụ (P)
# ========================================================
np.random.seed(42)
N = 500

# Giả lập 3 nhân tố ẩn có tương quan với nhau (Oblique structure)
latent_corr = np.array([[1.0, 0.4, -0.2], [0.4, 1.0, -0.1], [-0.2, -0.1, 1.0]])
factors = np.random.multivariate_normal([0, 0, 0], latent_corr, N)

data = {
    'T1': 0.8*factors[:,0] + np.random.normal(0, 0.4, N), 
    'T2': 0.7*factors[:,0] + np.random.normal(0, 0.4, N), 
    'T3': 0.85*factors[:,0] + np.random.normal(0, 0.4, N),
    'C1': 0.75*factors[:,1] + np.random.normal(0, 0.4, N), 
    'C2': 0.8*factors[:,1] + np.random.normal(0, 0.4, N), 
    'C3': 0.7*factors[:,1] + np.random.normal(0, 0.4, N),
    'P1': 0.9*factors[:,2] + np.random.normal(0, 0.4, N), 
    'P2': 0.85*factors[:,2] + np.random.normal(0, 0.4, N), 
    'P3': 0.8*factors[:,2] + np.random.normal(0, 0.4, N)
}
df = pd.DataFrame(data)

# ========================================================
# BƯỚC 2: KIỂM ĐỊNH SỰ PHÙ HỢP (KMO & BARTLETT)
# ========================================================
chi_square_value, p_value = calculate_bartlett_sphericity(df)
kmo_all, kmo_model = calculate_kmo(df)

print(f"---- KIỂM ĐỊNH SỰ PHÙ HỢP ----")
print(f"Bartlett's Test p-value: {p_value:.4f} ({'Đạt' if p_value < 0.05 else 'Không đạt'})")
print(f"Chỉ số KMO tổng thể   : {kmo_model:.4f} ({'Rất tốt' if kmo_model > 0.8 else 'Tạm được'})")

# ========================================================
# BƯỚC 3: THỰC HIỆN EFA VÀ XOAY NHÂN TỐ (PROMAX)
# ========================================================
fa = FactorAnalyzer(n_factors=3, rotation="promax")
fa.fit(df)

# Trích xuất Pattern Matrix (Ma trận Mẫu)
pattern_matrix = pd.DataFrame(fa.loadings_, 
                              index=df.columns, 
                              columns=['Factor_Trust', 'Factor_Convenience', 'Factor_Price'])

print("\n---- MA TRẬN MẪU (PATTERN MATRIX) SAU XOAY PROMAX ----")
display(pattern_matrix.round(3))
```

### Phân tích diễn giải Output:

1. **Kết quả Kiểm định:** 
   - KMO $\approx 0.85$ (> 0.6) và p-value của Bartlett < 0.05. Điều này xác nhận bộ dữ liệu khảo sát khách hàng ngân hàng này hoàn toàn đủ phẩm chất để phân tích nhân tố.
2. **Giải mã Pattern Matrix:**
   - Bạn có thể thấy `T1, T2, T3` chỉ tải mạnh vào `Factor_Trust` (> 0.6) và có hệ số rất thấp ở các nhân tố khác. 
   - Tương tự cho `Convenience` và `Price`. 
   - Việc các biến phân tách sạch sẽ vào các cột riêng biệt chính là biểu hiện của **Cấu trúc đơn giản (Simple Structure)**. Nhờ phép xoay Promax, ta có thể tự tin đặt tên cho các nhân tố ẩn mà không sợ bị nhiễu bởi sự chồng chéo giữa các khái niệm.

## 8.5. THỰC HÀNH PYTHON: XỬ LÝ DỮ LIỆU KHẢO SÁT VÀ KIỂM ĐỊNH THANG ĐO

Trong các nghiên cứu về Kinh doanh và Quản trị, chúng ta thường sử dụng bảng câu hỏi (Survey) để đo lường các khái niệm trừu tượng. Tuy nhiên, không phải câu hỏi nào cũng phản ánh tốt biến ẩn. Quy trình thực hành này bao gồm hai giai đoạn sống còn:
1. **Phân tích Nhân tố Khám phá (EFA):** Để kiểm tra giá trị hội tụ (Convergent Validity) và giá trị phân biệt (Discriminant Validity).
2. **Kiểm định Cronbach's Alpha:** Để đánh giá độ tin cậy (Reliability) hay tính nhất quán nội tại của các biến quan sát.

---

## 8.5.1. Cơ sở toán học của Độ tin cậy (Cronbach's Alpha)

Trước khi lập trình, chúng ta cần hiểu chỉ số $\alpha$ được tính toán như thế nào. Cronbach's Alpha đo lường mức độ tương quan giữa các biến trong cùng một nhân tố. Phương trình toán học:

$$\alpha = \frac{k}{k-1} \left( 1 - \frac{\sum_{i=1}^k \sigma_i^2}{\sigma_X^2} \right)$$

**Giải thích chi tiết các ký hiệu:**
- $k$: Số lượng biến quan sát (câu hỏi) trong một thang đo cụ thể.
- $\sigma_i^2$: Phương sai của biến quan sát thứ $i$.
- $\sigma_X^2$: Tổng phương sai của thang đo (tổng của tất cả các biến quan sát trong nhóm).
- **Ý nghĩa:** $\alpha$ dao động từ 0 đến 1. 
  - $\alpha > 0.8$: Thang đo rất tốt.
  - $0.7 < \alpha < 0.8$: Sử dụng được.
  - $0.6 < \alpha < 0.7$: Chấp nhận được trong các nghiên cứu mới.
  - $\alpha < 0.6$: Thang đo thiếu tin cậy, cần loại bỏ biến hoặc thiết kế lại câu hỏi.

---

```python
import numpy as np
import pandas as pd
from factor_analyzer import FactorAnalyzer
from factor_analyzer.factor_analyzer import calculate_kmo, calculate_bartlett_sphericity
import pingouin as pg
import seaborn as sns
import matplotlib.pyplot as plt

# ========================================================
# BƯỚC 1: GIẢ LẬP DỮ LIỆU KHẢO SÁT DỊCH VỤ NGÂN HÀNG
# Giả sử khảo sát 500 khách hàng về 3 nhóm yếu tố (Likert 1-5):
# - Sự tin tưởng (TR1, TR2, TR3, TR4)
# - Năng lực đáp ứng (RE1, RE2, RE3, RE4)
# - Cơ sở vật chất (TA1, TR2, TA3, TA4)
# ========================================================
np.random.seed(42)
N = 500

def generate_likert(base, noise_level=0.5):
    raw = base + np.random.normal(0, noise_level, N)
    return np.clip(np.round(raw), 1, 5).astype(int)

latent_trust = np.random.uniform(2, 4.5, N)
latent_resp = np.random.uniform(2, 4.5, N)
latent_tang = np.random.uniform(2, 4.5, N)

data = {
    'TR1': generate_likert(latent_trust), 'TR2': generate_likert(latent_trust), 
    'TR3': generate_likert(latent_trust), 'TR4': generate_likert(0.5 * latent_trust + 2), # Biến xấu 1: Loading thấp
    'RE1': generate_likert(latent_resp), 'RE2': generate_likert(latent_resp), 
    'RE3': generate_likert(latent_resp), 'RE4': generate_likert(0.6 * latent_trust + 0.6 * latent_resp), # Biến xấu 2: Tải chéo (Cross-loading)
    'TA1': generate_likert(latent_tang), 'TA2': generate_likert(latent_tang), 
    'TA3': generate_likert(latent_tang), 'TA4': generate_likert(latent_tang)
}

df = pd.DataFrame(data)
print("5 dòng dữ liệu khảo sát đầu tiên:")
display(df.head())
```

### 8.5.2. Kiểm định KMO và Bartlett

Trước khi chạy EFA, ta dùng `factor_analyzer` để xác nhận dữ liệu có đủ tính tương quan hay không.

```python
# 1. Kiểm định Bartlett
chi_square_value, p_value = calculate_bartlett_sphericity(df)
print(f"Bartlett's Test p-value: {p_value:.4f}")

# 2. Chỉ số KMO
kmo_all, kmo_model = calculate_kmo(df)
print(f"KMO Overall: {kmo_model:.4f}")

# Phân tích kết quả:
# Nếu p-value < 0.05 và KMO > 0.6, chúng ta có 'giấy thông hành' để thực hiện EFA.
```

### 8.5.3. Thực hiện EFA và Xoay Varimax

Chúng ta sẽ trích xuất nhân tố dựa trên trị riêng Eigenvalue > 1 và thực hiện xoay vuông góc (Varimax) để làm rõ ma trận hệ số tải.

```python
# Khởi tạo EFA
fa = FactorAnalyzer(rotation="varimax", n_factors=3, method='principal')
fa.fit(df)

# Trích xuất ma trận Loadings
loadings = pd.DataFrame(fa.loadings_, 
                           index=df.columns, 
                           columns=['Factor 1', 'Factor 2', 'Factor 3'])

print("---- MA TRẬN HỆ SỐ TẢI (FACTOR LOADINGS) ----")
display(loadings.round(3))

# Trực quan hóa bằng Heatmap
plt.figure(figsize=(8, 10))
sns.heatmap(loadings, annot=True, cmap='RdBu_r', center=0)
plt.title('Hình 8.1: Heatmap Hệ số tải nhân tố')
plt.show()
```

### 8.5.4. Chiến lược loại bỏ biến xấu

Dựa trên ma trận hệ số tải ở trên, một chuyên gia phân tích sẽ áp dụng các quy tắc loại bỏ biến (Item Deletion) như sau:

1. **Hệ số tải thấp (Low Loadings):** Bỏ các biến không tải được vào bất kỳ nhân tố nào với hệ số $> 0.3$. Nhìn vào `TR4`, nếu hệ số tải của nó ở cả 3 cột đều nhỏ, nó không đóng góp gì cho mô hình.
2. **Tải chéo (Cross-loadings):** Một biến tải mạnh vào hai nhân tố cùng lúc với độ chênh lệch hệ số tải $< 0.2$. Ví dụ `RE4` tải vào Factor 1 là 0.5 và Factor 2 là 0.45. Biến này gây nhiễu vì không biết nó thuộc nhóm nào.
3. **Biến rác (Uniqueness):** Biến có độ chung (Communality) $< 0.3$ (tương đương Uniqueness $> 0.7$).

### 8.5.5. Đánh giá độ tin cậy bằng Cronbach's Alpha

Sau khi đã xác định được nhóm các biến hội tụ vào từng nhân tố, chúng ta dùng thư viện `pingouin` để tính $\alpha$ cho từng nhóm.

```python
# Giả sử sau EFA ta xác định được 3 nhóm nhân tố:
factor_trust_items = ['TR1', 'TR2', 'TR3']
factor_resp_items = ['RE1', 'RE2', 'RE3']
factor_tang_items = ['TA1', 'TA2', 'TA3', 'TA4']

def check_reliability(items, name):
    # Tính Cronbach's Alpha
    alpha_res = pg.cronbach_alpha(data=df[items])
    alpha = alpha_res[0]
    
    print(f"Nhân tố {name}:")
    print(f" - Cronbach's Alpha: {alpha:.4f}")
    print(f" - Kết luận: {'Đạt chuẩn' if alpha >= 0.7 else 'Cần xem xét lại'}\n")

check_reliability(factor_trust_items, "Sự tin tưởng")
check_reliability(factor_resp_items, "Năng lực đáp ứng")
check_reliability(factor_tang_items, "Cơ sở vật chất")
```

### Phân tích diễn giải Output Báo cáo:

1. **Phân tích EFA:** 
   - Nhìn vào ma trận loadings, nếu các biến `TR1, TR2, TR3` cùng có hệ số dương lớn ở cột `Factor 1`, ta kết luận Factor 1 đại diện cho "Sự tin tưởng". 
   - Nếu biến `RE4` (biến chúng ta cố tình làm xấu) hiện lên với các hệ số tải sấp sỉ nhau ở hai cột, ta phải thực hiện lệnh `df.drop(columns=['RE4'])` và chạy lại EFA từ đầu. Đây là quá trình lặp (Iterative process) cho đến khi đạt được **Cấu trúc đơn giản (Simple Structure)**.

2. **Phân tích Cronbach's Alpha:**
   - Trong thực tế, nếu một nhóm biến có $\alpha$ thấp (ví dụ 0.55), hãy kiểm tra cột `Alpha if item deleted` (có sẵn trong hàm `pg.cronbach_alpha` nếu cấu hình chi tiết). Nếu việc xóa một biến cụ thể làm $\alpha$ vọt lên 0.7, đó là bằng chứng thép để loại bỏ biến đó khỏi thang đo chính thức.

**Kết luận cuối cùng:** Chỉ khi các nhân tố vượt qua cả kiểm định EFA (giá trị hội tụ và phân biệt) và Cronbach's Alpha (độ tin cậy), chúng ta mới đủ điều kiện tính giá trị trung bình nhân tố (Factor Mean) để đưa vào các mô hình hồi quy hoặc phân tích sự khác biệt (ANOVA) ở các chương sau.
