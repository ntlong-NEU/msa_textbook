# CHƯƠNG 3: HỒI QUY TUYẾN TÍNH ĐA BIẾN (MULTIPLE LINEAR REGRESSION)

Hồi quy tuyến tính đa biến là "ngựa thồ" (workhorse) của kinh tế lượng và phân tích dữ liệu tài chính. Trước khi tiến tới các thuật toán học máy phức tạp (Machine Learning) hay mạng nơ-ron (Neural Networks), việc thấu hiểu tường tận cơ chế hoạt động, phương pháp tối ưu hóa và các ranh giới giả định của mô hình tuyến tính là bắt buộc. Một mô hình tuyến tính được hiểu thấu đáo thường mang lại giá trị giải thích (Interpretability) vượt trội so với các mô hình hộp đen (Black-box models).

---

## 3.1. Mô hình Toán học OLS: Cực tiểu hóa Tổng bình phương phần dư

### 3.1.1. Phương trình Hồi quy Dạng Ma trận

Trong không gian đa biến, việc viết phương trình hồi quy dưới dạng tổng đại số cho từng quan sát là vô cùng cồng kềnh. Thay vào đó, chúng ta sử dụng ngôn ngữ của Đại số tuyến tính. Xét một tập dữ liệu gồm $N$ quan sát và $K$ tham số (bao gồm $K-1$ biến độc lập và 1 hệ số chặn). Phương trình hồi quy tổng thể (Population Regression Function) được biểu diễn bằng ma trận như sau:

$$\mathbf{Y} = \mathbf{X}\boldsymbol{\beta} + \boldsymbol{\epsilon}$$

**Giải thích chi tiết TỪNG ký hiệu toán học:**
- $\mathbf{Y}$: Vector cột kích thước $(N \times 1)$, chứa các giá trị thực tế của biến phụ thuộc (Dependent variable). Ví dụ: Lợi suất của $N$ cổ phiếu.
- $\mathbf{X}$: Ma trận thiết kế (Design Matrix) kích thước $(N \times K)$. Cột đầu tiên của $\mathbf{X}$ là một vector toàn số 1 (để nhân với hệ số chặn $\beta_0$). Các cột còn lại từ $2$ đến $K$ chứa giá trị của các biến độc lập (Explanatory variables). Đây là dữ liệu đầu vào đã biết.
- $\boldsymbol{\beta}$: Vector cột kích thước $(K \times 1)$, chứa các hệ số hồi quy (Regression coefficients) cần phải ước lượng. $\beta_k$ cho biết sự thay đổi biên của $\mathbf{Y}$ khi biến $\mathbf{X}_k$ tăng 1 đơn vị, trong điều kiện các biến khác không đổi (Ceteris paribus).
- $\boldsymbol{\epsilon}$: Vector cột kích thước $(N \times 1)$, chứa các sai số ngẫu nhiên (Random error/Disturbance terms). Nó đại diện cho tất cả các yếu tố phi tuyến, các biến bị bỏ sót, hoặc các cú sốc ngẫu nhiên tác động lên $\mathbf{Y}$ mà $\mathbf{X}$ không thể giải thích được.

### 3.1.2. Chứng minh Toán học: Tìm nghiệm $\hat{\boldsymbol{\beta}}$ bằng phương pháp OLS

Mục tiêu của phương pháp Bình phương tối thiểu thông thường (Ordinary Least Squares - OLS) là tìm một vector ước lượng $\hat{\boldsymbol{\beta}}$ sao cho tổng bình phương của các phần dư (Residual Sum of Squares - RSS) là nhỏ nhất. Phần dư $\mathbf{e}$ là sự khác biệt giữa giá trị thực tế và giá trị dự báo: $\mathbf{e} = \mathbf{Y} - \hat{\mathbf{Y}} = \mathbf{Y} - \mathbf{X}\hat{\boldsymbol{\beta}}$.

Hàm mục tiêu RSS được định nghĩa là tích vô hướng của vector phần dư với chính nó:

$$\text{RSS} = \mathbf{e}^T \mathbf{e} = (\mathbf{Y} - \mathbf{X}\hat{\boldsymbol{\beta}})^T (\mathbf{Y} - \mathbf{X}\hat{\boldsymbol{\beta}})$$

**Bước 1: Khai triển hàm đa thức ma trận**
Áp dụng quy tắc chuyển vị $(\mathbf{A}\mathbf{B})^T = \mathbf{B}^T\mathbf{A}^T$, ta có:
$$\text{RSS} = (\mathbf{Y}^T - \hat{\boldsymbol{\beta}}^T\mathbf{X}^T) (\mathbf{Y} - \mathbf{X}\hat{\boldsymbol{\beta}})$$
$$\text{RSS} = \mathbf{Y}^T\mathbf{Y} - \mathbf{Y}^T\mathbf{X}\hat{\boldsymbol{\beta}} - \hat{\boldsymbol{\beta}}^T\mathbf{X}^T\mathbf{Y} + \hat{\boldsymbol{\beta}}^T\mathbf{X}^T\mathbf{X}\hat{\boldsymbol{\beta}}$$

Lưu ý quan trọng: Tích $\hat{\boldsymbol{\beta}}^T\mathbf{X}^T\mathbf{Y}$ tạo ra một đại lượng vô hướng (scalar) có kích thước $(1 \times 1)$. Vì chuyển vị của một số vô hướng chính là chính nó, nên $\hat{\boldsymbol{\beta}}^T\mathbf{X}^T\mathbf{Y} = (\hat{\boldsymbol{\beta}}^T\mathbf{X}^T\mathbf{Y})^T = \mathbf{Y}^T\mathbf{X}\hat{\boldsymbol{\beta}}$.

Gộp hai số hạng giữa lại, ta được hàm mục tiêu rút gọn:
$$\text{RSS} = \mathbf{Y}^T\mathbf{Y} - 2\hat{\boldsymbol{\beta}}^T\mathbf{X}^T\mathbf{Y} + \hat{\boldsymbol{\beta}}^T\mathbf{X}^T\mathbf{X}\hat{\boldsymbol{\beta}}$$

**Bước 2: Lấy đạo hàm bậc nhất (Vector Calculus)**
Để tìm cực tiểu của hàm $\text{RSS}$ theo vector $\hat{\boldsymbol{\beta}}$, ta tính Gradient (đạo hàm riêng theo từng phần tử của $\hat{\boldsymbol{\beta}}$) và cho bằng vector không $\mathbf{0}$.
Áp dụng các quy tắc vi tích phân ma trận:
1. $\frac{\partial (\mathbf{a}^T \mathbf{x})}{\partial \mathbf{x}} = \mathbf{a}$ $\Rightarrow \frac{\partial (\hat{\boldsymbol{\beta}}^T\mathbf{X}^T\mathbf{Y})}{\partial \hat{\boldsymbol{\beta}}} = \mathbf{X}^T\mathbf{Y}$
2. $\frac{\partial (\mathbf{x}^T \mathbf{A} \mathbf{x})}{\partial \mathbf{x}} = 2\mathbf{A}\mathbf{x}$ (với $\mathbf{A}$ là ma trận đối xứng) $\Rightarrow \frac{\partial (\hat{\boldsymbol{\beta}}^T(\mathbf{X}^T\mathbf{X})\hat{\boldsymbol{\beta}})}{\partial \hat{\boldsymbol{\beta}}} = 2\mathbf{X}^T\mathbf{X}\hat{\boldsymbol{\beta}}$

Đạo hàm của RSS:
$$\frac{\partial \text{RSS}}{\partial \hat{\boldsymbol{\beta}}} = \mathbf{0} - 2\mathbf{X}^T\mathbf{Y} + 2\mathbf{X}^T\mathbf{X}\hat{\boldsymbol{\beta}} = \mathbf{0}$$

**Bước 3: Giải Hệ phương trình chuẩn (Normal Equations)**
Chia cả hai vế cho 2 và chuyển vế, ta thu được phương trình hạt nhân của OLS:
$$\mathbf{X}^T\mathbf{X}\hat{\boldsymbol{\beta}} = \mathbf{X}^T\mathbf{Y}$$

Nhân cả hai vế (từ bên trái) với ma trận nghịch đảo của $(\mathbf{X}^T\mathbf{X})$, ký hiệu là $(\mathbf{X}^T\mathbf{X})^{-1}$, ta có nghiệm cuối cùng:

$$\hat{\boldsymbol{\beta}} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{Y}$$

*(Lưu ý: Để chắc chắn đây là điểm cực tiểu chứ không phải cực đại, ta lấy đạo hàm bậc hai (Ma trận Hessian): $\frac{\partial^2 \text{RSS}}{\partial \hat{\boldsymbol{\beta}} \partial \hat{\boldsymbol{\beta}}^T} = 2\mathbf{X}^T\mathbf{X}$. Vì $\mathbf{X}^T\mathbf{X}$ là một ma trận bán xác định dương (Positive Semi-Definite), hàm RSS là một hàm lồi chặt (strictly convex), do đó nghiệm $\hat{\boldsymbol{\beta}}$ vừa tìm được là điểm Cực tiểu Toàn cục (Global Minimum).)*

## 3.2. Định lý Gauss-Markov & Các giả định kinh điển

Có vô số cách để vẽ một đường thẳng qua một đám mây điểm. Tại sao OLS lại được coi là "tiêu chuẩn vàng"? Câu trả lời nằm ở **Định lý Gauss-Markov**, được phát biểu rằng: Nếu 5 giả định kinh điển của OLS được thỏa mãn, thì bộ trọng số $\hat{\boldsymbol{\beta}}$ của OLS chính là **BLUE (Best Linear Unbiased Estimator)**.

**Giải mã khái niệm BLUE:**
- **Linear (Tuyến tính):** $\hat{\boldsymbol{\beta}}$ là một tổ hợp tuyến tính của biến phụ thuộc $\mathbf{Y}$ (do $\hat{\boldsymbol{\beta}} = \mathbf{W}\mathbf{Y}$ với $\mathbf{W} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T$).
- **Unbiased (Không chệch):** Kỳ vọng toán học của ước lượng đúng bằng giá trị thực của tổng thể: $E[\hat{\boldsymbol{\beta}}] = \boldsymbol{\beta}$. Mô hình không nhắm lệch mục tiêu.
- **Best (Tốt nhất):** Trong số tất cả các ước lượng tuyến tính không chệch, OLS có **phương sai nhỏ nhất** (Minimum Variance). Các khoảng tin cậy (Confidence Intervals) của OLS là hẹp nhất, mang lại độ chính xác cao nhất.

Tuy nhiên, danh hiệu BLUE đi kèm với một cái giá rất đắt. Nó yêu cầu dữ liệu phải tuân thủ nghiêm ngặt 5 giả định sau:

### Giả định 1: Tuyến tính trong các tham số (Linearity in Parameters)
Mô hình $Y = \beta_0 + \beta_1 X_1 + \dots + \beta_k X_k + \epsilon$ phải là hàm bậc nhất đối với các hệ số $\beta$. 
*Chú ý:* Biến độc lập $\mathbf{X}$ có thể phi tuyến (ví dụ $X_1^2, \ln(X_2)$), nhưng các tham số $\beta$ tuyệt đối không được phi tuyến (như $\beta_1^2$ hay $e^{\beta_2}$). Nếu vi phạm, ta không thể dùng ma trận $(\mathbf{X}^T\mathbf{X})^{-1}$ để giải.

### Giả định 2: Không có đa cộng tuyến hoàn hảo (No Perfect Multicollinearity)
Rank của ma trận $\mathbf{X}$ phải bằng $K$ (Full column rank). Về mặt đại số, điều này đảm bảo định thức $\det(\mathbf{X}^T\mathbf{X}) \neq 0$, và ma trận này có thể nghịch đảo được.
*Rủi ro thực tế:* Trong quản trị rủi ro, nếu bạn đưa cả biến "Tổng dư nợ VND" và biến "Tổng dư nợ quy đổi USD" vào mô hình (với một tỷ giá cố định), sự đa cộng tuyến là hoàn hảo 100%. Máy tính sẽ báo lỗi Singular Matrix và hệ thống sụp đổ vì có vô số lời giải cho $\boldsymbol{\beta}$.

### Giả định 3: Kỳ vọng phần dư có điều kiện bằng 0 (Zero Conditional Mean)
$$E[\boldsymbol{\epsilon} | \mathbf{X}] = 0$$
Sai số ngẫu nhiên $\epsilon$ không được chứa bất kỳ thông tin nào liên quan đến $\mathbf{X}$. Điều này đảm bảo rằng phần dư thực sự là "nhiễu trắng" (white noise). 
*Sự sụp đổ (Endogeneity):* Nếu ta muốn đánh giá tác động của "Ngân sách Marketing" lên "Doanh thu", nhưng ta bỏ sót biến "Chất lượng sản phẩm" (Omitted Variable). Vì "Chất lượng" tương quan với cả Marketing và Doanh thu, phần dư $\epsilon$ lúc này chứa thông tin của "Chất lượng" và có tương quan dương với $\mathbf{X}$. Hậu quả: $\hat{\beta}$ sẽ bị chệch (Biased), phóng đại quá mức sức mạnh của Marketing.

### Giả định 4: Phương sai sai số không đổi (Homoskedasticity)
$$Var(\boldsymbol{\epsilon} | \mathbf{X}) = \sigma^2 \mathbf{I}$$
Độ phân tán của sai số phải là một hằng số $\sigma^2$ cho mọi mức độ của $\mathbf{X}$. Mọi điểm trên đường hồi quy đều có mức độ "bất định" như nhau.
*Hậu quả khi vi phạm (Heteroskedasticity):* Xảy ra phổ biến trong dữ liệu tài chính (ví dụ: công ty quy mô lớn thì biến động lợi nhuận có biên độ lớn hơn hẳn công ty nhỏ). Khi phương sai dạng phễu xuất hiện, $\hat{\boldsymbol{\beta}}$ vẫn không chệch, nhưng nó **mất danh hiệu "Best"**. Các sai số chuẩn (Standard Errors) tính bằng OLS sẽ bị sai, dẫn đến việc các phép kiểm định t-test (tính p-value) trở nên vô giá trị, làm ta kết luận sai về sự có ý nghĩa của biến.

### Giả định 5: Không có hiện tượng tự tương quan (No Autocorrelation)
$$Cov(\epsilon_i, \epsilon_j | \mathbf{X}) = 0 \quad \forall i \neq j$$
Phần dư của quan sát này không được dự báo phần dư của quan sát khác. Giả định này thường bị vi phạm trong dữ liệu chuỗi thời gian (Time-series), ví dụ lạm phát tháng này cao do các cú sốc vĩ mô, thì lạm phát tháng sau nhiều khả năng cũng sẽ cao (Tự tương quan dương). Vi phạm giả định này cũng phá hủy tính hợp lệ của Standard Errors và p-value.

### Case Study & Lập trình Python: Định giá Bất động sản

Chúng ta sẽ kiểm chứng sức mạnh của ma trận OLS và kiểm tra giả định Homoskedasticity thông qua một tập dữ liệu giả lập về định giá căn hộ dựa trên Diện tích (Square footage) và Tuổi đời (Age).

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Khởi tạo dữ liệu giả lập (N = 300 căn hộ)
np.random.seed(42)
N = 300
Area = np.random.uniform(50, 200, N)        # Diện tích từ 50m2 đến 200m2
Age = np.random.uniform(0, 30, N)           # Tuổi đời từ 0 đến 30 năm

# Thiết lập mô hình tổng thể thực (True Population Model): 
# Price = 1000 + 20*Area - 15*Age + Error
# Giả lập Heteroskedasticity: Căn hộ càng lớn, phương sai giá (phần dư) càng rộng
error_variance = (Area / 20) ** 2  # Phương sai tỷ lệ thuận với bình phương diện tích
errors = np.random.normal(0, np.sqrt(error_variance), N)

Price = 1000 + 20 * Area - 15 * Age + errors

# Tạo DataFrame
df = pd.DataFrame({'Price': Price, 'Area': Area, 'Age': Age})

# ==========================================
# BƯỚC 1: TÍNH TÓAN THỦ CÔNG MA TRẬN OLS
# ==========================================
# Xây dựng ma trận X (thêm cột hằng số 1)
X_matrix = np.column_stack((np.ones(N), Area, Age))
Y_vector = Price.values

# Áp dụng công thức Beta = (X^T * X)^-1 * X^T * Y
XT_X = X_matrix.T @ X_matrix
XT_X_inv = np.linalg.inv(XT_X)
XT_Y = X_matrix.T @ Y_vector
beta_hat = XT_X_inv @ XT_Y

print("---- KẾT QUẢ ƯỚC LƯỢNG OLS THỦ CÔNG (ĐẠI SỐ TUYẾN TÍNH) ----")
print(f"Hệ số chặn (Beta_0)   : {beta_hat[0]:.4f}")
print(f"Hệ số Diện tích (Beta_1): {beta_hat[1]:.4f}")
print(f"Hệ số Tuổi đời (Beta_2) : {beta_hat[2]:.4f}\n")

# ==========================================
# BƯỚC 2: SỬ DỤNG STATSMODELS ĐỂ KIỂM ĐỊNH
# ==========================================
X_sm = sm.add_constant(df[['Area', 'Age']])
model = sm.OLS(df['Price'], X_sm).fit()
print("---- KIỂM CHỨNG BẰNG THƯ VIỆN STATSMODELS ----")
print(model.summary().tables[1])

# Lấy phần dư từ mô hình
residuals = model.resid
```

```python
# ==========================================
# BƯỚC 3: TRỰC QUAN HÓA KIỂM ĐỊNH GAUSS-MARKOV (HETEROSKEDASTICITY)
# ==========================================
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Area'], y=residuals, color='navy', alpha=0.6)
plt.axhline(0, color='red', linestyle='--', linewidth=2)
plt.title('Đồ thị Phần dư theo Diện tích (Residuals vs Fitted)\nHiện tượng Phương sai thay đổi (Heteroskedasticity)', fontsize=14)
plt.xlabel('Diện tích căn hộ (m2)')
plt.ylabel('Phần dư (Residuals - Sai số dự báo)')
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()
```

### Phân tích Output Python:
1. **Sức mạnh của Toán học Ma trận:** Kết quả `beta_hat` tính bằng phép nhân ma trận $(X^T X)^{-1} X^T Y$ qua `numpy` khớp chính xác 100% với output của thư viện `statsmodels` chuyên nghiệp. Mô hình xấp xỉ lại rất tốt hàm thực tế (Hệ số chặn ~1000, Area ~20, Age ~-15).
2. **Phá vỡ giả định Gauss-Markov:** Hãy nhìn vào biểu đồ phần dư (Residual Plot). Tại các mức diện tích nhỏ (50-100m2), phần dư bám rất sát đường 0 (phương sai nhỏ, định giá rất chính xác). Tuy nhiên, khi căn hộ lên tới 150-200m2, phần dư bung rộng ra như hình một cái phễu (phương sai cực lớn, sai số định giá có thể lên tới hàng chục nghìn USD). 
   - Hiện tượng này chính là **Heteroskedasticity (Phương sai thay đổi)** vi phạm nghiêm trọng giả định số 4. 
   - *Cách khắc phục:* Chuyên gia phân tích sẽ phải sử dụng Standard Errors dạng Robust (như White's Robust Standard Errors - `cov_type='HC3'` trong python) hoặc thực hiện phép chuyển đổi Logarit trên biến giá nhà (Log-transform) để ép cái phễu này co lại thành hình ống chuẩn (Homoskedasticity).

# 3.3. Đánh giá mô hình & Suy diễn thống kê (Statistical Inference)

Sau khi tìm được vector hệ số ước lượng $\hat{\boldsymbol{\beta}}$ bằng OLS, câu hỏi đặt ra là: Lời giải này đáng tin cậy đến mức nào? Mô hình của chúng ta giải thích được bao nhiêu phần trăm sự thật, và liệu các hệ số tìm được có ý nghĩa thống kê hay chỉ là kết quả của sự ngẫu nhiên trong việc lấy mẫu?

### 3.3.1. Hệ số xác định $R^2$ và $R^2$ hiệu chỉnh (Adjusted $R^2$)

**1. Hệ số xác định $R^2$ (Coefficient of Determination):**
Để đo lường "độ khít" (Goodness-of-fit) của mô hình, chúng ta sử dụng $R^2$. Nó trả lời câu hỏi: Sự biến thiên của các biến độc lập $\mathbf{X}$ giải thích được bao nhiêu phần trăm sự biến thiên của biến phụ thuộc $Y$?

Về mặt toán học, tổng biến thiên của $Y$ (Total Sum of Squares - TSS) được chia thành 2 phần: Biến thiên do mô hình giải thích được (Explained Sum of Squares - ESS) và Biến thiên không giải thích được (Residual Sum of Squares - RSS).

$$\text{TSS} = \sum_{i=1}^n (y_i - \bar{y})^2$$
$$\text{ESS} = \sum_{i=1}^n (\hat{y}_i - \bar{y})^2$$
$$\text{RSS} = \sum_{i=1}^n (y_i - \hat{y}_i)^2 = \sum_{i=1}^n e_i^2$$
$$\text{TSS} = \text{ESS} + \text{RSS}$$

Hệ số $R^2$ được định nghĩa là:
$$R^2 = \frac{\text{ESS}}{\text{TSS}} = 1 - \frac{\text{RSS}}{\text{TSS}}$$

**2. Sự lừa dối của $R^2$ và sự ra đời của Adjusted $R^2$:**
$R^2$ có một lỗ hổng chí mạng: Nó **không bao giờ giảm** khi ta thêm một biến độc lập mới vào mô hình, kể cả khi biến đó hoàn toàn vô nghĩa (ví dụ: dự báo giá cổ phiếu bằng chu kỳ mặt trăng). Toán học đằng sau điều này rất đơn giản: OLS luôn tìm cách tối thiểu hóa RSS. Việc có thêm một biến (thêm một bậc tự do cho tối ưu hóa) sẽ chỉ làm RSS giữ nguyên hoặc giảm xuống, từ đó ép $R^2$ tăng lên. Điều này khuyến khích hành vi "nhồi nhét" biến (Overfitting).

Để trừng phạt hành vi nhồi nhét này, ta sử dụng **Adjusted $R^2$** (Ký hiệu: $\bar{R}^2$). Nó hiệu chỉnh $R^2$ bằng cách xét đến số bậc tự do (Degrees of freedom):

$$\bar{R}^2 = 1 - \left( \frac{\text{RSS} / (N - k - 1)}{\text{TSS} / (N - 1)} \right) = 1 - (1 - R^2)\frac{N - 1}{N - k - 1}$$

**Giải thích ký hiệu:**
- $N$: Kích thước mẫu (Số lượng quan sát).
- $k$: Số lượng biến độc lập.
Khi thêm một biến mới ($k$ tăng), mẫu số $(N - k - 1)$ giảm, tạo ra một "hình phạt" làm giảm $\bar{R}^2$. $\bar{R}^2$ chỉ tăng khi và chỉ khi biến mới được thêm vào làm giảm RSS đủ mạnh để bù đắp lại hình phạt của việc mất đi một bậc tự do.

---

### 3.3.2. F-test (Kiểm định sự phù hợp toàn cục) và t-test (Kiểm định ý nghĩa từng hệ số)

**1. F-test: Mô hình có giá trị hay không?**
Trước khi phân tích từng biến, ta phải kiểm tra xem toàn bộ mô hình có tốt hơn việc dùng một giá trị trung bình $\bar{y}$ để dự báo hay không.
- **Giả thuyết Không ($H_0$):** $\beta_1 = \beta_2 = \dots = \beta_k = 0$ (Tất cả các biến độc lập đều không có tác động đến $Y$).
- **Giả thuyết Đối ($H_A$):** Có ít nhất một $\beta_j \neq 0$.
- **Thống kê F:** 
$$F = \frac{\text{ESS} / k}{\text{RSS} / (N - k - 1)} = \frac{R^2 / k}{(1 - R^2) / (N - k - 1)}$$
Nếu F-statistic > Giá trị F tới hạn (hoặc p-value < 0.05), ta bác bỏ $H_0$ và kết luận mô hình có ý nghĩa thống kê tổng thể.

**2. t-test: Biến này có thực sự quan trọng?**
Nếu F-test pass, ta soi chiếu vào từng biến độc lập $X_j$.
- **Giả thuyết Không ($H_0$):** $\beta_j = 0$ (Biến $X_j$ không có tác động biên lên $Y$, khi đã kiểm soát các biến khác).
- **Thống kê t:** 
$$t = \frac{\hat{\beta}_j - 0}{\text{SE}(\hat{\beta}_j)}$$
- **Giải thích:** $\text{SE}(\hat{\beta}_j)$ là Sai số chuẩn (Standard Error) của hệ số $\beta_j$. Thống kê t cho biết giá trị ước lượng $\hat{\beta}_j$ nằm cách $0$ bao nhiêu lần sai số chuẩn. Thông thường, nếu $|t| > 1.96$ (tương ứng với mức ý nghĩa 5% trong phân phối chuẩn), ta tự tin 95% rằng $\beta_j$ thực sự khác 0.

# 3.4. Omitted Variable Bias (Thiên lệch do bỏ sót biến)

Một trong những câu thần chú quan trọng nhất trong thống kê là: *"Tương quan không ngụ ý nhân quả"* (Correlation does not imply causation). Nguyên nhân lớn nhất dẫn đến sự nhầm lẫn này trong hồi quy tuyến tính chính là Omitted Variable Bias (OVB).

### 3.4.1. Bản chất Toán học của OVB

Giả sử **Mô hình thực (True Model)** của thế giới (nhưng ta không biết) được chi phối bởi hai biến $X_1$ và $X_2$:
$$Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \epsilon$$

Nhưng vì giới hạn dữ liệu, ta không thu thập được $X_2$. Ta chạy một **Mô hình thiếu (Omitted Model)** chỉ với $X_1$:
$$Y = \tilde{\beta}_0 + \tilde{\beta}_1 X_1 + u$$
Trong đó, phần dư mới $u$ giờ đây đã nuốt trọn biến $X_2$: $u = \beta_2 X_2 + \epsilon$.

Theo công thức OLS đơn biến, hệ số ước lượng $\tilde{\beta}_1$ được tính bằng:
$$\tilde{\beta}_1 = \frac{Cov(X_1, Y)}{Var(X_1)}$$

Thay phương trình thế giới thực của $Y$ vào tử số:
$$\tilde{\beta}_1 = \frac{Cov(X_1, \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \epsilon)}{Var(X_1)}$$

Áp dụng tính chất phân phối của Hiệp phương sai:
$$\tilde{\beta}_1 = \beta_1 \frac{Var(X_1)}{Var(X_1)} + \beta_2 \frac{Cov(X_1, X_2)}{Var(X_1)} + \frac{Cov(X_1, \epsilon)}{Var(X_1)}$$

Với giả định $\epsilon$ là nhiễu trắng ($Cov(X_1, \epsilon) = 0$), lấy kỳ vọng hai vế ta thu được **Định lý OVB kinh điển**:
$$E[\tilde{\beta}_1] = \beta_1 + \beta_2 \frac{Cov(X_1, X_2)}{Var(X_1)}$$

Đặt $\delta_{21} = \frac{Cov(X_1, X_2)}{Var(X_1)}$. (Chú ý: $\delta_{21}$ chính là hệ số dốc nếu ta lấy $X_2$ hồi quy theo $X_1$).

**Độ lệch (Bias)** của mô hình là:
$$\text{Bias} = E[\tilde{\beta}_1] - \beta_1 = \beta_2 \times \delta_{21}$$

**Điều kiện sống còn:** Mô hình OLS sẽ bị thiên lệch trầm trọng NẾU VÀ CHỈ NẾU cả hai điều kiện sau đồng thời xảy ra:
1. $\beta_2 \neq 0$: Biến bị bỏ sót ($X_2$) thực sự có ảnh hưởng đến $Y$.
2. $\delta_{21} \neq 0$: Biến bị bỏ sót ($X_2$) có sự tương quan (covariance $\neq 0$) với biến đang có trong mô hình ($X_1$).

### 3.4.2. Case Study Kinh tế: Phương trình thu nhập Mincer (Mincer Earnings Function)

Đây là bài toán nổi tiếng nhất của Kinh tế học lao động (Labor Economics). Bạn là nhà hoạch định chính sách muốn đo lường mức độ tác động của **Số năm học vấn ($X_1$)** lên **Tiền lương ($Y$)** để quyết định ngân sách đầu tư cho giáo dục.

Tuy nhiên, có một biến số ẩn cực kỳ quan trọng không bao giờ đo lường chính xác được: **Năng lực bẩm sinh/Chỉ số IQ ($X_2$)**.

Phân tích dấu của OVB:
- $\beta_2$ (Tác động của IQ lên Lương): Chắc chắn **> 0** (Người thông minh, nhạy bén hơn thường có thu nhập cao hơn bất kể bằng cấp).
- $\delta_{21}$ (Tương quan giữa IQ và Số năm học): Chắc chắn **> 0** (Người có tư duy tốt thường dễ dàng học lên Cao học, Tiến sĩ hơn; ngược lại, hệ thống giáo dục cũng có xu hướng đào thải những người có năng lực tiếp thu kém).

=> **Bias = (+) $\times$ (+) = Positive Bias (Thiên lệch dương).**

**Kết luận chấn động:** Nếu bạn chỉ chạy hồi quy Lương theo Số năm học, hệ số $\hat{\beta}_1$ bạn tìm được sẽ bị **phóng đại (overestimated)** so với giá trị thật. Nó không chỉ phản ánh phần thưởng của việc cắp sách đến trường, mà nó đã "ăn cắp" luôn cả tín hiệu của trí thông minh bẩm sinh hòa vào trong đó.

Dưới đây là phần mô phỏng bằng Python để chứng minh toán học này.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

# -----------------------------------------------------
# 1. MÔ PHỎNG DỮ LIỆU THẾ GIỚI THỰC (TRUE DATA GENERATING PROCESS)
# -----------------------------------------------------
np.random.seed(123)
N = 1000

# X2: Năng lực bẩm sinh (Ability/IQ), giả sử phân phối chuẩn N(100, 15)
Ability = np.random.normal(100, 15, N)

# X1: Số năm học vấn (Education). TƯƠNG QUAN DƯƠNG VỚI ABILITY.
# Người có Ability cao thường học lâu hơn.
Education = 5 + 0.08 * Ability + np.random.normal(0, 1.5, N)

# Y: Tiền lương (Wage). Chịu tác động của cả Education (Beta1=50) và Ability (Beta2=20)
Wage = 500 + 50 * Education + 20 * Ability + np.random.normal(0, 150, N)

df = pd.DataFrame({'Wage': Wage, 'Education': Education, 'Ability': Ability})

# -----------------------------------------------------
# 2. CHẠY MÔ HÌNH THỰC (TRUE MODEL) CÓ ĐẦY ĐỦ BIẾN
# -----------------------------------------------------
X_true = sm.add_constant(df[['Education', 'Ability']])
model_true = sm.OLS(df['Wage'], X_true).fit()
print("========== MÔ HÌNH THỰC (TRUE MODEL: KHÔNG BỊ BỎ SÓT) ==========")
print(f"Hệ số Education thực sự (True Beta_1) : {model_true.params['Education']:.2f}")
print(f"Hệ số Ability thực sự (True Beta_2)   : {model_true.params['Ability']:.2f}\n")

# -----------------------------------------------------
# 3. CHẠY MÔ HÌNH THIẾU (OMITTED MODEL) - BỎ SÓT ABILITY
# -----------------------------------------------------
X_omitted = sm.add_constant(df[['Education']])
model_omitted = sm.OLS(df['Wage'], X_omitted).fit()
print("========== MÔ HÌNH BỊ THIÊN LỆCH (OMITTED VARIABLE BIAS) ==========")
print(f"Hệ số Education ảo (Biased Beta_1)    : {model_omitted.params['Education']:.2f}")

# -----------------------------------------------------
# 4. KIỂM CHỨNG CÔNG THỨC TOÁN HỌC OVB
# -----------------------------------------------------
# Tính Delta_21 (Hồi quy X2 theo X1)
delta_model = sm.OLS(df['Ability'], sm.add_constant(df['Education'])).fit()
delta_21 = delta_model.params['Education']
beta_2_true = model_true.params['Ability']
beta_1_true = model_true.params['Education']

calculated_biased_beta1 = beta_1_true + (beta_2_true * delta_21)

print("\n========== CHỨNG MINH CÔNG THỨC TOÁN HỌC ==========")
print(f"Độ lệch (Bias) = Beta_2 * Delta_21 = {beta_2_true:.2f} * {delta_21:.2f} = {beta_2_true * delta_21:.2f}")
print(f"Beta_1 tính theo lý thuyết OVB = {beta_1_true:.2f} + {(beta_2_true * delta_21):.2f} = {calculated_biased_beta1:.2f}")
print(f"=> Khớp hoàn toàn với Beta_1 chạy từ mô hình thiếu ({model_omitted.params['Education']:.2f})!")
```

### Phân tích diễn giải Output:

Khi chạy đoạn mã trên, bạn sẽ quan sát thấy một sự thật phũ phàng của phân tích dữ liệu:
1. **Mô hình thực:** Tác động biên thuần túy của 1 năm học lên mức lương chỉ là **~50 đơn vị** (Trùng khớp với kỳ vọng khai báo `Beta1 = 50`).
2. **Mô hình thiếu:** Khi ta ngây thơ chạy hồi quy chỉ với biến Giáo dục (do không đo lường được IQ), hệ số này nhảy vọt lên thành **~220 đơn vị**!
3. **Tác hại:** Nếu lấy kết quả 220 này để tư vấn cho Chính phủ, bạn đang truyền đạt một thông điệp sai lầm rằng "giáo dục tạo ra giá trị kinh khủng", trong khi thực tế 3/4 giá trị đó đến từ trí thông minh tự thân của con người. 

Sự hiểu biết về OVB giúp các chuyên gia dữ liệu luôn phải hoài nghi và đặt câu hỏi: *"Còn yếu tố nào chưa được kiểm soát đang giật dây cả biến độc lập lẫn biến phụ thuộc không?"* trước khi đưa ra bất kỳ kết luận nhân quả nào.

# 3.5. Bệnh lý của mô hình: Đa cộng tuyến (Multicollinearity)

Trong lý thuyết hồi quy, một mô hình lý tưởng là khi các biến độc lập (Predictors) có tương quan mạnh với biến phụ thuộc (Target), nhưng lại **hoàn toàn độc lập (trực giao) với nhau**. Tuy nhiên, trong kinh tế học, điều này gần như không tồn tại. Lạm phát luôn đi kèm với lãi suất, diện tích nhà luôn đi kèm với số phòng ngủ. 

Khi sự phụ thuộc tuyến tính giữa các biến độc lập trở nên quá lớn, mô hình mắc phải căn bệnh **Đa cộng tuyến (Multicollinearity)**.

### 3.5.1. Bản chất toán học và Hệ số phóng đại phương sai (VIF)

Hãy nhớ lại phương trình tính Phương sai của hệ số ước lượng $\hat{\beta}_j$ trong hồi quy đa biến:

$$Var(\hat{\beta}_j) = \frac{\sigma^2}{(N-1)Var(X_j)} \times \frac{1}{1 - R_j^2}$$

**Giải phẫu công thức:**
- $\sigma^2$: Phương sai của sai số ngẫu nhiên (Nhiễu của hệ thống).
- $(N-1)Var(X_j)$: Tổng bình phương độ lệch của biến $X_j$. Kích thước mẫu càng lớn hoặc dữ liệu của $X_j$ càng phân tán rộng, phương sai của $\hat{\beta}_j$ càng nhỏ (ước lượng càng chính xác).
- **$R_j^2$ (Hệ số xác định phụ):** Đây là phần quan trọng nhất. $R_j^2$ là $R^2$ thu được khi ta lấy biến $X_j$ làm biến phụ thuộc và **hồi quy nó theo tất cả các biến độc lập còn lại** trong mô hình.

Từ đây, ta định nghĩa **Hệ số phóng đại phương sai (Variance Inflation Factor - VIF)**:

$$VIF_j = \frac{1}{1 - R_j^2}$$

Đại lượng này cho biết phương sai của hệ số $\hat{\beta}_j$ đã bị "bơm phồng" (inflated) lên bao nhiêu lần do hiện tượng đa cộng tuyến so với khi $X_j$ hoàn toàn độc lập trực giao với các biến khác.

### 3.5.2. Hậu quả của VIF cao

Khi $X_j$ có thể được dự báo gần như hoàn hảo bởi các biến khác (VD: $R_j^2 = 0.9$), thì $VIF_j = 1 / (1 - 0.9) = 10$. Điều gì sẽ xảy ra?

1. **Bùng nổ sai số chuẩn (Standard Errors Inflation):** Phương sai tăng gấp 10 lần đồng nghĩa với việc Sai số chuẩn (SE) tăng $\sqrt{10} \approx 3.16$ lần. Khoảng tin cậy của $\hat{\beta}_j$ sẽ trở nên rộng mênh mông.
2. **Tê liệt kiểm định t (Insignificant t-tests):** Vì $t = \hat{\beta}_j / SE$, khi SE quá lớn, thống kê t sẽ trượt về gần 0, khiến p-value > 0.05. Kết quả là biến $X_j$ bị kết luận là "không có ý nghĩa thống kê", mặc dù thực tế nó rất quan trọng.
3. **Hiện tượng "Dấu phi logic":** Thuật toán OLS sẽ bị bối rối trong việc phân bổ sự đóng góp. Nó có thể gán một hệ số âm vô lý cho một biến rõ ràng phải có tác động dương (VD: Số phòng ngủ tăng nhưng giá nhà lại giảm).
4. **Mô hình mong manh (Fragile Model):** Chỉ cần thêm hoặc bớt một vài quan sát nhỏ lẻ, các hệ số $\beta$ có thể lộn nhào từ dương sang âm, hoặc thay đổi độ lớn một cách kịch liệt.

### 3.5.3. Ngưỡng cảnh báo và Cách khắc phục
- **Quy tắc ngón tay cái (Rule of Thumb):** 
  - VIF < 5: Chấp nhận được (Đa cộng tuyến nhẹ).
  - 5 < VIF < 10: Cảnh báo nguy hiểm, cần xem xét mô hình.
  - VIF > 10: Đa cộng tuyến nghiêm trọng, bắt buộc phải xử lý.
- **Giải pháp:**
  1. *Loại bỏ biến (Feature Drop):* Bỏ đi một trong các biến mang thông tin trùng lặp (VD: Đã có "Tổng diện tích" thì cân nhắc bỏ "Diện tích tầng 1").
  2. *Gộp biến (Index creation):* Cộng gộp các biến tương quan thành một chỉ số (Index) duy nhất.
  3. *Giảm chiều (PCA):* Chuyển đổi dữ liệu sang không gian trực giao bằng Principal Component Analysis.
  4. *Hồi quy có hình phạt (Regularization):* Sử dụng Ridge Regression (L2 penalty) để kìm hãm sự bùng nổ của các hệ số.

# 3.6. Thực hành Python: Chẩn đoán OLS với Statsmodels

Chúng ta sẽ xây dựng một mô hình dự báo Giá Bất Động Sản. Để minh họa bệnh lý, tôi sẽ cố tình đưa vào các biến có tính Đa cộng tuyến cực mạnh (Diện tích và Số phòng ngủ), sau đó sẽ đọc "bệnh án" thông qua bảng summary().

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------------------------------
# BƯỚC 1: TẠO DỮ LIỆU BẤT ĐỘNG SẢN GIẢ LẬP (N=500)
# -----------------------------------------------------
np.random.seed(42)
N = 500

# Biến 1: Diện tích (Area) từ 30m2 đến 200m2
Area = np.random.uniform(30, 200, N)

# Biến 2: Số phòng (Rooms) - Phụ thuộc tuyến tính RẤT MẠNH vào Area (Đa cộng tuyến)
Rooms = np.round(Area / 30 + np.random.normal(0, 0.5, N))

# Biến 3: Khoảng cách tới trung tâm (Distance_km)
Distance = np.random.uniform(1, 25, N)

# Biến Phụ thuộc: Giá nhà (Price - Tỷ VND)
# Phương trình thực: Price = 1.0 + 0.05*Area + 0.0*Rooms - 0.1*Distance + Nhiễu
# Chú ý: Trong thực tế, Rooms không đóng góp thêm giá trị nếu đã biết Area.
Price = 1.0 + 0.05 * Area + 0.0 * Rooms - 0.1 * Distance + np.random.normal(0, 0.5, N)

df = pd.DataFrame({'Price': Price, 'Area': Area, 'Rooms': Rooms, 'Distance': Distance})

# -----------------------------------------------------
# BƯỚC 2: KIỂM TRA HỆ SỐ VIF TRƯỚC KHI HỒI QUY
# -----------------------------------------------------
X_vars = df[['Area', 'Rooms', 'Distance']]
X_with_const = sm.add_constant(X_vars)

vif_data = pd.DataFrame()
vif_data["Feature"] = X_with_const.columns
vif_data["VIF"] = [variance_inflation_factor(X_with_const.values, i) 
                   for i in range(len(X_with_const.columns))]

print("BẢNG CHẨN ĐOÁN VIF (VARIANCE INFLATION FACTOR):")
print(vif_data.round(2))
print("\n=> Nhận xét: VIF của Area và Rooms đều xấp xỉ 10, dấu hiệu nguy hiểm của Đa cộng tuyến!\n")

# -----------------------------------------------------
# BƯỚC 3: HUẤN LUYỆN MÔ HÌNH VÀ IN BÁO CÁO
# -----------------------------------------------------
model = sm.OLS(df['Price'], X_with_const).fit()
print(model.summary())
```

### Giải nghĩa chuyên sâu TỪNG CHỈ SỐ trong Bảng OLS Summary

Khi chạy đoạn mã trên, Python sẽ trả về một bảng báo cáo gồm 3 phần chính. Đây là cách một chuyên gia kinh tế lượng đọc hiểu nó:

**Phần 1: Thông tin tổng quan (Overall Model Fit)**
- **Dep. Variable (Biến phụ thuộc):** `Price`.
- **R-squared ($R^2$):** (Ví dụ: 0.850). Mô hình giải thích được 85% sự biến thiên của giá nhà.
- **Adj. R-squared:** Luôn nhỏ hơn hoặc bằng $R^2$, đã phạt yếu tố nhồi nhét biến. Nếu $R^2$ và Adj. $R^2$ lệch nhau quá nhiều, mô hình của bạn đang chứa nhiều biến vô dụng.
- **F-statistic & Prob (F-statistic):** Kiểm định ý nghĩa tổng thể. Nếu Prob (p-value) < 0.05, ta kết luận có ít nhất một biến X ảnh hưởng đến Y. Trong ví dụ này, nó bằng $0.00$, mô hình hoàn toàn hợp lệ.
- **Log-Likelihood / AIC / BIC:** Dùng để so sánh các mô hình với nhau. AIC/BIC càng nhỏ, mô hình càng tối ưu (đạt sự cân bằng giữa độ khít và độ đơn giản).

**Phần 2: Bảng hệ số (Coefficient Table) - Bị ảnh hưởng bởi Đa cộng tuyến**
- **coef (Hệ số $\hat{\beta}$):** Cho biết tác động biên. VD: `Distance` có coef $\approx -0.1$. Khi nhà cách xa trung tâm thêm 1km, giá giảm 0.1 tỷ VND.
- **std err (Sai số chuẩn):** Đây là "độ rung lắc" của $\hat{\beta}$. Hãy nhìn vào `Area` và `Rooms`. Vì VIF $\approx 10$ (Đa cộng tuyến), std err của chúng bị bơm phồng rất lớn.
- **t & P>|t| (Kiểm định t và p-value):** 
  - Mặc dù `Area` là yếu tố cốt lõi quyết định giá nhà (trong quá trình tạo dữ liệu), nhưng do sự có mặt của `Rooms`, hệ thống bị nhiễu. 
  - Tùy vào mẫu ngẫu nhiên, bạn có thể thấy p-value của `Rooms` hoặc thậm chí `Area` > 0.05. Mô hình thất bại trong việc xác định ai mới là kẻ thực sự đóng góp giá trị. Đây chính là hệ quả tàn khốc của Đa cộng tuyến.
- **[0.025      0.975] (Khoảng tin cậy 95%):** Nếu khoảng này chứa số 0 (chạy từ âm sang dương), biến đó không có ý nghĩa thống kê.

**Phần 3: Kiểm định phần dư (Residual Diagnostics)**
- **Omnibus & Prob(Omnibus):** Kiểm định tính chuẩn (Normality) của phần dư. Giả định Gauss-Markov không ép phần dư phải chuẩn, nhưng nếu chuẩn, các kiểm định t và F sẽ chính xác hơn ở mẫu nhỏ. Nếu Prob > 0.05, phần dư tuân theo phân phối chuẩn.
- **Skewness (Độ lệch) & Kurtosis (Độ nhọn):** Phân phối chuẩn hoàn hảo có Skew = 0 và Kurtosis = 3.
- **Durbin-Watson:** Kiểm định tự tương quan (Autocorrelation) bậc 1. Giá trị nằm trong khoảng $[0, 4]$.
  - DW $\approx 2.0$: Tốt! Không có tự tương quan.
  - DW tiến về 0: Tự tương quan dương (Thường gặp trong chuỗi thời gian).
  - DW tiến về 4: Tự tương quan âm.
- **Jarque-Bera (JB):** Một kiểm định khác cho Normality (kết hợp Skewness và Kurtosis).
- **Cond. No. (Condition Number):** Chỉ số báo động Đa cộng tuyến cuối cùng. Bất kỳ giá trị nào > 30 đều ám chỉ ma trận $(X^T X)$ đang gần suy biến do đa cộng tuyến nặng.

```python
# -----------------------------------------------------
# BƯỚC 4: VẼ BIỂU ĐỒ PHẦN DƯ (RESIDUAL PLOTS)
# Kiểm tra giả định Phương sai không đổi (Homoskedasticity)
# -----------------------------------------------------
fitted_vals = model.fittedvalues
residuals = model.resid

fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# 1. Biểu đồ Fitted vs Residuals
sns.scatterplot(x=fitted_vals, y=residuals, alpha=0.6, ax=axes[0], color='dodgerblue')
axes[0].axhline(0, color='red', linestyle='--', linewidth=2)
axes[0].set_title('Hình 3.1: Residuals vs Fitted Values', fontsize=14)
axes[0].set_xlabel('Giá nhà dự báo (Fitted Values)')
axes[0].set_ylabel('Phần dư (Residuals)')
axes[0].grid(True, linestyle=':', alpha=0.6)

# 2. Biểu đồ Q-Q Plot (Kiểm tra phân phối chuẩn của phần dư)
sm.qqplot(residuals, line='45', fit=True, ax=axes[1], color='dodgerblue', alpha=0.6)
axes[1].set_title('Hình 3.2: Normal Q-Q Plot', fontsize=14)
axes[1].grid(True, linestyle=':', alpha=0.6)

plt.tight_layout()
plt.show()
```

### Phân tích Đồ thị Chẩn đoán (Diagnostic Plots)

1. **Hình 3.1 (Residuals vs Fitted Values):**
   - Mục tiêu: Kiểm tra xem lỗi dự báo (phần dư) có phân tán đều đặn ở mọi mức giá nhà hay không.
   - Nếu các điểm dữ liệu tạo thành một dải băng hình chữ nhật với bề rộng không đổi xung quanh đường 0 (như kết quả mô phỏng trên), mô hình thỏa mãn giả định **Homoskedasticity (Phương sai không đổi)**.
   - Nếu các điểm tạo thành hình cái phễu (mở rộng về phía tay phải), hoặc có dạng đường cong (Parabol), mô hình đang bị thiếu biến hoặc cần chuyển đổi Logarit trên biến Y.

2. **Hình 3.2 (Q-Q Plot):**
   - Mục tiêu: Trực quan hóa phân phối của phần dư so với một phân phối chuẩn lý tưởng.
   - Nếu đa số các điểm xanh nằm bám sát vào đường thẳng đỏ (góc 45 độ), phần dư phân phối chuẩn (Omnibus an toàn). Nếu 2 đầu đuôi văng ra xa, dữ liệu có hiện tượng đuôi béo (Fat-tails) hoặc chứa nhiều Outliers.
